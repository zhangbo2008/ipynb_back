```python
!pip install diffusers >>/dev/null
import torch
from diffusers import StableDiffusionPipeline

pipe = StableDiffusionPipeline.from_pretrained("CompVis/stable-diffusion-v1-4", revision="fp16", torch_dtype=torch.float16)
pipe = pipe.to("cuda")
prompt = ["a photograph of an astronaut riding a horse"]*3
tmp=pipe(prompt, guidance_scale=7.5, num_inference_steps=15,)
from IPython.display import display


for i in range(len(tmp.nsfw_content_detected)):
    
    print(tmp.nsfw_content_detected[i],'打印是否生成非法图片')
    display(tmp.images[i])
    
```
